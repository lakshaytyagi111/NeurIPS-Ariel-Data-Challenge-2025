```python
import numpy as np

# Load your downloaded preprocessed data
signal = np.load("/kaggle/input/signal-v2-1-npy/signal_v2 (1).npy")

print("Shape:", signal.shape)
print("Data type:", signal.dtype)
```

```python
planet_idx = 0  # change to inspect others

fgs1_signal = signal[planet_idx, 0]     # (T,)  — single detector
airs_signal = signal[planet_idx, 1]     # (T, F) — spectral detector

print("FGS1 shape:", fgs1_signal.shape)
print("AIRS-CH0 shape:", airs_signal.shape)

```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.plot(fgs1_signal)
plt.title("FGS1 Flux vs Time")
plt.xlabel("Time bin")
plt.ylabel("Flux")
plt.show()
```

```python
for idx in range(3):
    plt.plot(signal[idx, 0], label=f'Planet {idx}')
plt.title("FGS1 Light Curves (first 3 planets)")
plt.legend()
plt.show()
```
